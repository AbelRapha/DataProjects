```python
#Utilizando a Biblioteca Matplotlib
import matplotlib.pyplot as plt
#Criação dos eixos x e y

x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]

#Criação do Gráfico

plt.plot(x,y, color = "blue") #Gráfico de linha
plt.scatter(x,y, color = "red") #Criação dos pontos vermelhos
plt.title(" Dados de Demanda")
plt.xlabel("Dia")
plt.ylabel("Produtos")
plt.grid()#Criação da grade
plt.legend(["Estimativa", "Dado"])#Criação da legenda do ponto e linha
plt.show()
```

<h1><strong>Variações no Gráfico</strong></h1>

```python
#Criacao dos eixos

x = list(range(1,10,3))
y = list(range(10,19, 3))

plt.plot(x,y,"p-.")
plt.scatter(x,y, marker = "*", color = "red")
plt.show()

```

<strong>Legendas para o comando plot </strong>

```python
# Criação dos eixos

x  = list(range(0,10,2))
y1 = list(range(10,20,2))
y2 = list(range(10,15,1))

plt.plot(x,y1, color = "blue", label = "Produção")
plt.plot(x,y2,color = "red", label = "Demanda")
plt.legend()
plt.grid()
plt.show()
```

<h2><strong> Gráficos de Linha com Gráficos de Barra</strong></h2>

```python
#Criação dos eixos

x  = list(range(0,10,2))
y1 = list(range(10,20,2))
y2 = list(range(10,15,1))

plt.plot(x,y1, "r.-")
plt.bar(x,y2, color = "gray")
plt.show()
```

<h2><strong>Dois gráficos de barras juntos</setrog></h2>

```python
#Defenindo as coordenadas para cada variável

x1 = list(range(1,10,1))
x2 = list(range(5,15,1))
y1 = list(range(1,10,1))
y2 = list(range(10,20,1))

plt.bar(x1,y1,color = "blue", label = "População 1")
plt.bar(x2,y2,color = "gray", label = "População 2")
plt.legend()
plt.show()
```
